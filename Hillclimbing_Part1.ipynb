```python
import math
#initialize x as -3
x = -3
#initialize stepsize to test value
stepsize = 0.01
#while x is greater than or equal to -5 and less than or equal to 5 perform hillclimbing
while x >= -5 and x <= 5:
  #if x - stepsize is a greater number when put into the function than x, decrement x by stepsize
  if (2 - pow(x,2) < 2 - pow(x-stepsize, 2)):
    print("Subtracting: ", stepsize, " Original: ", 2 - pow(x,2), " < Step: " , pow(x,2), " Changing x to : ", x - stepsize);
    x -= stepsize
  #if x + stepsize is a greater number when put into the function than x increment x by stepsize
  elif(2 - pow(x,2) < 2 - pow(x+stepsize,2)):
    print("Adding: ", stepsize, " Original: ", 2 - pow(x,2), " < Step: " , 2 - pow(x+stepsize,2), " Changing x to : ", x + stepsize);
    x += stepsize
  else:
    #if neither of those are true then a maximum has been found so break the loop
    break
#print the maximum
print(x);

```
