```python
import math
import random

#Function for equation to avoid having to type this out several times and cluttering the code
def equation(x):
  return ((0.0051 * pow(x,5)) - (0.1367 * pow(x,4)) + (1.24 * pow(x,3)) - (4.456 * pow(x,2)) + (5.66 * x) - 0.287)
#initialize stepsize
stepsize = 0.5
#create variable max value to store the x value with the greatest y value when put into the equation
bestX = 0;
#loop 20 times for the 20 different random starts
for i in range(20):
  #set the seed to i so that it is random each iteration comment out for hillclimbing
  random.seed(i)
  #set the random x value as a number between 0 and 10 comment out for hillclimbing
  x = random.uniform(0,10)
  #For the hillclimbing part set x = 3
  #x = 3
  #make sure x is between the bounds of 0 and 10 when in the hillclimbing loop
  while x >= 0 and x <= 10:
    #if x-stepsize yields a greater y value then decrement x by stepsize
    if (equation(x) < equation(x - stepsize)):
      print("Subtracting: ", stepsize, " Original: ", equation(x), " < Step: " , equation(x - stepsize), " Changing x to : ", x - stepsize);
      x -= stepsize
    #elseif x+stepsize yields a greater y value then increment x by stepsize
    elif(equation(x) < equation(x + stepsize)):
      print("Adding: ", stepsize, " Original: ", equation(x), " < Step: " , equation(x + stepsize), " Changing x to : ", x + stepsize);
      x += stepsize
    #if y is larger with x than steps in either direction set x
    else:
      #check if the y value with x is better than the other version of x if so set this x to the new bestX
      if equation(x) >= equation(bestX):
        print("New Max Value: ", x, "Changing from ", bestX)
        bestX = x
      #break the loop to grab a new random x value
      break
#print the final x value with the largest y value when plugged into the equation
print(bestX)
```
